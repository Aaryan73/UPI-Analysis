## Exploratory Analysis and Visualization



### Plot showing the increase in amount of UPI transaction in billions over the course of 42 months

```python
plt.title("Amount of transactions in billion over the course of 42 months")
plt.xlabel('Time | 0 being the latest date')
plt.ylabel('Transaction in millions')
plt.plot(UPI_df["Amount.Of.Transaction.In.Bn"]);
```

### Plot showing the increase in volume of UPI transaction in millions over the course of 42 months

```python
plt.title("Volume of transactions in millions over the course of 42 monthsr")
plt.xlabel('Time | 0 being the latest date')
plt.ylabel('Transaction in millions')
plt.plot(UPI_df["Volume.Of.Transaction.in.Mn"]);
```

### The graph below shows a jump in amount because the data collected is not continuous. We are looking at the difference after a year 

```python
plt.title("Amunt of transactions in billion per year")
plt.xlabel('Year')
plt.ylabel('Transaction in millions')
sns.scatterplot(UPI_df["Year"], UPI_df["Amount.Of.Transaction.In.Bn"]);
plt.plot(UPI_df["Year"], UPI_df["Amount.Of.Transaction.In.Bn"], "o--g");
```

## Asking and Answering Questions

### What is UPI?
(Wikipedia Definition)
-> Unified Payments Interface (UPI) is an instant real-time payment system developed by National Payments Corporation of India facilitating inter-bank transactions. The interface is regulated by the Reserve Bank of India and works by instantly transferring funds between two bank accounts on a mobile platform.

### How is the growth rate in transaction amount over the course of 42 months?

```python
inc_rate = (UPI_df.at[0, "Amount.Of.Transaction.In.Bn"] - UPI_df.at[42, "Amount.Of.Transaction.In.Bn"])/UPI_df.at[42, "Amount.Of.Transaction.In.Bn"]
```

```python
inc_rate
```

-> The amount of UPI transactions in India has increased by almost 7200 times in just 42 months or 3.5 years. It is really a remarkable progress and other countries should also take it as an inspiration. 

### What is the increase in number of banks using UPI?

```python
number_increased = UPI_df.at[0, "No.Of.Banks.Using.Upi"] - UPI_df.at[42, "No.Of.Banks.Using.Upi"]
```

```python
print("The number of banks using UPI has incresed by {} in india over the course of 42 months.".format(number_increased))
```

### How can India improve this progress to even higher level?
-> Looking at the scatter-plot of "Number of banks using UPI" and "Transaction amount" (fifth graph), I think it is fair enough to say that the amount of transaction increases exponentially as the number of banks using UPI increases.
